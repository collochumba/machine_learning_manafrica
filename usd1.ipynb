```python
#want to focus now on the aple stock.
aapl = close.AAPL.copy().to_frame()
```

```python
aapl
```

```python
aapl.shift( periods = 1) #
```

```python
aapl["lag1"] = aapl.shift(periods = 1) # this creates another column in dataframe of aapl
```

```python
aapl
```

```python
aapl["Diff"] = aapl.AAPL.sub(aapl.lag1) #(Diff) that shows day-to-day price movement
```

```python
aapl
```

```python
aapl["% change"] = aapl.AAPL.div(aapl.lag1).sub(1).mul(100)
aapl
```

```python
aapl["% change 2"] = aapl.AAPL.pct_change(periods = 1).mul(100) # make the percentage change into readable percentages
aapl
```

```python
del aapl["lag1"]
del aapl["Diff"]
del aapl["% change"] # deleting columns
```

```python
aapl
```

```python
aapl.rename( columns = {"% change 2" : "change"}, inplace = True)
```

```python
aapl
```

```python
aapl.AAPL.resample("M").last() # gives monthly
aapl
```

```python
aapl.AAPL.resample("BM").last()# gives last business day
aapl
```

```python
aapl.AAPL.resample("BME").last().pct_change(periods = 1).mul(100) # gives percentage change last business day
aapl
```

```python
ret = aapl.pct_change().dropna()
```

```python
ret 
```

```python
ret.replace([float("inf"), float("-inf")], float("nan"), inplace=True)
ret = ret.dropna()
ret.plot(kind="hist", figsize=(12,8), bins=100)
plt.show()
```

```python
ret['AAPL'] = ret['AAPL'].apply(lambda x: np.log(x+1) if x > 0 else x)
ret['change'] = ret['change'].apply(lambda x: np.log(abs(x)+1) if x != 0 else x)
ret.plot(kind="hist", figsize=(12,8), bins=100)
plt.show()
```

```python
ax = ret['AAPL'].plot(kind="hist", figsize=(12,8), bins=100, color="blue", alpha=0.5)
ret['change'].plot(kind="hist", bins=100, color="orange", alpha=0.5, ax=ax)
ax.set_ylabel('Frequency')
ax.set_title('Stock Price and Change Histogram')
plt.legend(["AAPL", "change"])
plt.show()
```
