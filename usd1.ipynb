```python
ab = normclose.plot(figsize=(12,6), fontsize = 12)
ab.set_title("Normalized Close data for AAPL SPY KO")
ab.set_xlabel("Date")
ab.set_ylabel("Price")
plt.legend(fontsize = 12)
plt.show()
```

```python
#want to focus now on the aple stock.
aapl = close.AAPL.copy().to_frame()
```

```python
aapl
```

```python
aapl.shift( periods = 1) #
```

```python
aapl["lag1"] = aapl.shift(periods = 1) # this creates another column in dataframe of aapl
```

```python
aapl
```

```python
aapl["Diff"] = aapl.AAPL.sub(aapl.lag1) #(Diff) that shows day-to-day price movement
```

```python
aapl
```

```python
aapl["% change"] = aapl.AAPL.div(aapl.lag1).sub(1).mul(100)
aapl
```

```python
aapl["% change 2"] = aapl.AAPL.pct_change(periods = 1).mul(100) # make the percentage change into readable percentages
aapl
```

```python
del aapl["lag1"]
del aapl["Diff"]
del aapl["% change"] # deleting columns
```

```python
aapl
```

```python
aapl.rename( columns = {"% change 2" : "change"}, inplace = True)
```

```python
aapl
```

```python
aapl.AAPL.resample("M").last() # gives monthly
aapl
```

```python
aapl.AAPL.resample("BM").last()# gives last business day
aapl
```

```python
aapl.AAPL.resample("BME").last().pct_change(periods = 1).mul(100) # gives percentage change last business day
aapl
```
