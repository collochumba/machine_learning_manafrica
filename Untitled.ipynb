```python
# ELITE PROFESSIONAL BETTING SYSTEM - HEDGE FUND GRADE
# All critical issues fixed + portfolio optimization + CLV tracking

import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')

from scipy.stats import poisson
from scipy.optimize import minimize
from scipy.special import softmax

from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import log_loss
from sklearn.calibration import CalibratedClassifierCV
import xgboost as xgb

import matplotlib.pyplot as plt
from tqdm import tqdm

print("🚀 ELITE SYSTEM - LOADING...")


# ============================================================================
# PHASE 1: ADVANCED FEATURE ENGINEERING
# ============================================================================

def create_elite_features(df):
    """
    Elite feature set with zero leakage.
    
    Adds:
    - Shots & shots on target
    - Corners
    - ELO ratings
    - Head-to-head records
    """
    df = df.copy().sort_values(['League', 'Date']).reset_index(drop=True)
    
    print("🔧 Engineering elite features...")
    
    # Basic rolling features (goals) - SHIFT FIRST
    for window in [5, 10]:
        df[f'HGS_L{window}'] = df.groupby(['League', 'HomeTeam'])['FTHG'].transform(
            lambda x: x.shift(1).rolling(window, min_periods=1).mean()
        )
        df[f'AGS_L{window}'] = df.groupby(['League', 'AwayTeam'])['FTAG'].transform(
            lambda x: x.shift(1).rolling(window, min_periods=1).mean()
        )
    
    # Advanced: Shots (if available)
    if 'HS' in df.columns and 'AS' in df.columns:
        for window in [5]:
            df[f'HShots_L{window}'] = df.groupby(['League', 'HomeTeam'])['HS'].transform(
                lambda x: x.shift(1).rolling(window, min_periods=1).mean()
            )
            df[f'AShots_L{window}'] = df.groupby(['League', 'AwayTeam'])['AS'].transform(
                lambda x: x.shift(1).rolling(window, min_periods=1).mean()
            )
    
    # Advanced: Shots on target (if available)
    if 'HST' in df.columns and 'AST' in df.columns:
        for window in [5]:
            df[f'HSOT_L{window}'] = df.groupby(['League', 'HomeTeam'])['HST'].transform(
                lambda x: x.shift(1).rolling(window, min_periods=1).mean()
            )
            df[f'ASOT_L{window}'] = df.groupby(['League', 'AwayTeam'])['AST'].transform(
                lambda x: x.shift(1).rolling(window, min_periods=1).mean()
            )
    
    # Advanced: Corners (if available)
    if 'HC' in df.columns and 'AC' in df.columns:
        for window in [5]:
            df[f'HCorners_L{window}'] = df.groupby(['League', 'HomeTeam'])['HC'].transform(
                lambda x: x.shift(1).rolling(window, min_periods=1).mean()
            )
            df[f'ACorners_L{window}'] = df.groupby(['League', 'AwayTeam'])['AC'].transform(
                lambda x: x.shift(1).rolling(window, min_periods=1).mean()
            )
    
    # ELO ratings (simplified)
    df['HomeELO'] = 1500.0
    df['AwayELO'] = 1500.0
    
    for league in df['League'].unique():
        league_df = df[df['League'] == league].copy()
        elo_ratings = {}
        
        for idx, row in league_df.iterrows():
            home = row['HomeTeam']
            away = row['AwayTeam']
            
            # Get current ELO
            home_elo = elo_ratings.get(home, 1500)
            away_elo = elo_ratings.get(away, 1500)
            
            df.at[idx, 'HomeELO'] = home_elo
            df.at[idx, 'AwayELO'] = away_elo
            
            # Update ELO after match
            if row['FTR'] == 'H':
                result = 1.0
            elif row['FTR'] == 'D':
                result = 0.5
            else:
                result = 0.0
            
            expected = 1 / (1 + 10 ** ((away_elo - home_elo) / 400))
            K = 30
            
            home_elo_new = home_elo + K * (result - expected)
            away_elo_new = away_elo + K * ((1 - result) - (1 - expected))
            
            elo_ratings[home] = home_elo_new
            elo_ratings[away] = away_elo_new
    
    # Form
    df['HP'] = df.apply(lambda x: 3 if x['FTR']=='H' else (1 if x['FTR']=='D' else 0), axis=1)
    df['AP'] = df.apply(lambda x: 3 if x['FTR']=='A' else (1 if x['FTR']=='D' else 0), axis=1)
    
    df['HForm'] = df.groupby(['League', 'HomeTeam'])['HP'].transform(
        lambda x: x.shift(1).rolling(5, min_periods=1).sum()
    )
    df['AForm'] = df.groupby(['League', 'AwayTeam'])['AP'].transform(
        lambda x: x.shift(1).rolling(5, min_periods=1).sum()
    )
    
    # League dummies
    league_dummies = pd.get_dummies(df['League'], prefix='Lg')
    df = pd.concat([df, league_dummies], axis=1)
    
    # Get feature columns
    feature_cols = [c for c in df.columns if any(x in c for x in ['_L', 'Form', 'ELO', 'Lg_'])]
    df = df.dropna(subset=feature_cols)
    
    print(f"✅ Created {len(feature_cols)} elite features")
    
    return df, feature_cols


# ============================================================================
# PHASE 2: TIME-AWARE CALIBRATED ML (FIXED)
# ============================================================================

def train_elite_ml(train_df, feature_cols):
    """
    FIXED: Uses TimeSeriesSplit inside calibration.
    
    This is CRITICAL for proper walk-forward validation.
    """
    X = train_df[feature_cols].fillna(0).values
    y = train_df['Outcome'].values
    
    print("🤖 Training elite calibrated ML...")
    
    # Base XGBoost
    base_model = xgb.XGBClassifier(
        objective='multi:softprob',
        num_class=3,
        max_depth=5,
        learning_rate=0.05,
        n_estimators=300,
        subsample=0.8,
        colsample_bytree=0.8,
        random_state=42,
        verbosity=0
    )
    
    # CRITICAL FIX: Time-aware CV
    tscv = TimeSeriesSplit(n_splits=5)
    
    calibrated_model = CalibratedClassifierCV(
        base_model,
        method='isotonic',
        cv=tscv  # ← FIXED: Time-aware, not random
    )
    
    calibrated_model.fit(X, y)
    
    # Evaluate
    y_pred = calibrated_model.predict_proba(X)
    ll = log_loss(y, y_pred)
    
    print(f"✅ Elite ML trained - Log-loss: {ll:.4f}")
    
    return calibrated_model


# ============================================================================
# PHASE 3: EXACT POISSON DIXON-COLES (NO MONTE CARLO)
# ============================================================================

class EliteDixonColes:
    """
    Elite Dixon-Coles with exact Poisson grid.
    
    FIX: No Monte Carlo noise - deterministic probabilities.
    """
    
    def __init__(self, xi=0.002):
        self.xi = xi
        self.teams = None
        self.attack = None
        self.defence = None
        self.home_adv = None
        self.rho = None
    
    def fit(self, df, league=None):
        if league:
            data = df[df['League'] == league].copy()
        else:
            data = df.copy()
        
        self.teams = sorted(set(data['HomeTeam'].unique()) | set(data['AwayTeam'].unique()))
        n = len(self.teams)
        
        # Vectorized
        team_to_idx = {team: i for i, team in enumerate(self.teams)}
        home_idx = data['HomeTeam'].map(team_to_idx).values
        away_idx = data['AwayTeam'].map(team_to_idx).values
        hg = data['FTHG'].values
        ag = data['FTAG'].values
        days = data['DaysSinceMatch'].values
        weights = np.exp(-self.xi * days)
        
        def nll(params):
            att = params[:n]
            deff = params[n:2*n]
            home = params[2*n]
            rho = params[2*n+1]
            
            lh = np.exp(home + att[home_idx] - deff[away_idx])
            la = np.exp(att[away_idx] - deff[home_idx])
            
            p = poisson.pmf(hg, lh) * poisson.pmf(ag, la)
            
            # Rho correction
            corr = np.ones_like(p)
            mask00 = (hg == 0) & (ag == 0)
            mask01 = (hg == 0) & (ag == 1)
            mask10 = (hg == 1) & (ag == 0)
            mask11 = (hg == 1) & (ag == 1)
            
            corr[mask00] = 1 - lh[mask00] * la[mask00] * rho
            corr[mask01] = 1 + lh[mask01] * rho
            corr[mask10] = 1 + la[mask10] * rho
            corr[mask11] = 1 - rho
            
            p *= corr
            ll = np.sum(weights * np.log(np.maximum(p, 1e-10)))
            return -ll
        
        x0 = np.concatenate([np.zeros(n), np.zeros(n), [0.2], [0.0]])
        res = minimize(nll, x0, method='L-BFGS-B', options={'maxiter': 100, 'ftol': 1e-5})
        
        self.attack = dict(zip(self.teams, res.x[:n]))
        self.defence = dict(zip(self.teams, res.x[n:2*n]))
        self.home_adv = res.x[2*n]
        self.rho = res.x[2*n+1]
        
        return self
    
    def predict_exact(self, home, away, max_goals=10):
        """
        FIXED: Exact Poisson grid calculation.
        
        No Monte Carlo - deterministic and fast.
        """
        lh = np.exp(self.home_adv + self.attack[home] - self.defence[away])
        la = np.exp(self.attack[away] - self.defence[home])
        
        # Exact Poisson grid
        probs = np.zeros((max_goals, max_goals))
        
        for i in range(max_goals):
            for j in range(max_goals):
                probs[i, j] = poisson.pmf(i, lh) * poisson.pmf(j, la)
        
        # Calculate outcomes
        prob_home = np.sum(probs[i, j] for i in range(max_goals) for j in range(max_goals) if i > j)
        prob_draw = np.sum(probs[i, j] for i in range(max_goals) for j in range(max_goals) if i == j)
        prob_away = np.sum(probs[i, j] for i in range(max_goals) for j in range(max_goals) if i < j)
        
        # Over/Under 2.5
        prob_over_25 = np.sum(probs[i, j] for i in range(max_goals) for j in range(max_goals) if (i + j) > 2.5)
        
        return {
            'lambda_home': float(lh),
            'lambda_away': float(la),
            'prob_home': float(prob_home),
            'prob_draw': float(prob_draw),
            'prob_away': float(prob_away),
            'prob_over_25': float(prob_over_25)
        }


# ============================================================================
# PHASE 4: FIXED LOG-ODDS ENSEMBLE
# ============================================================================

def ensemble_fixed(dc_probs, ml_probs, dc_weight=0.6):
    """
    FIXED: Independent log-odds, not relative to baseline.
    
    This is mathematically cleaner.
    """
    # Clip
    dc_probs = np.clip(dc_probs, 1e-9, 1-1e-9)
    ml_probs = np.clip(ml_probs, 1e-9, 1-1e-9)
    
    # Independent logs (FIXED)
    dc_log = np.log(dc_probs)
    ml_log = np.log(ml_probs)
    
    # Blend
    combined_log = dc_weight * dc_log + (1 - dc_weight) * ml_log
    
    # Softmax
    ensemble = softmax(combined_log)
    
    return ensemble


# ============================================================================
# PHASE 5: CLV TRACKER
# ============================================================================

class CLVTracker:
    """
    Closing Line Value tracker.
    
    Measures if you beat the closing line.
    """
    
    def __init__(self):
        self.clvs = []
    
    def calculate_clv(self, opening_odds, closing_odds):
        """
        CLV = closing_implied - opening_implied
        
        Positive CLV = you got better odds than closing.
        """
        if pd.isna(closing_odds) or closing_odds <= 1.0:
            return 0.0
        
        opening_implied = 1.0 / opening_odds
        closing_implied = 1.0 / closing_odds
        
        clv = (closing_implied - opening_implied) / opening_implied
        
        return clv
    
    def record(self, clv):
        self.clvs.append(clv)
    
    def get_stats(self):
        if not self.clvs:
            return {'mean_clv': 0, 'positive_pct': 0}
        
        return {
            'mean_clv': np.mean(self.clvs) * 100,
            'positive_pct': (np.array(self.clvs) > 0).mean() * 100
        }


# ============================================================================
# PHASE 6: MULTI-MARKET PORTFOLIO
# ============================================================================

def find_all_value_bets(ensemble_probs, odds, min_ev=0.03, min_prob=0.55):
    """
    FIXED: Find ALL value bets, not just best one.
    
    Real portfolios use multiple edges.
    """
    value_bets = []
    
    # Margin-adjusted EV
    implied_probs = 1 / odds
    overround = implied_probs.sum()
    true_market_probs = implied_probs / overround
    
    for i in range(3):
        prob = ensemble_probs[i]
        odd = odds[i]
        
        if prob < min_prob:  # Confidence filter
            continue
        
        edge = prob - true_market_probs[i]
        ev = (prob * odd) - 1
        
        if ev > min_ev:  # EV filter
            value_bets.append({
                'market_idx': i,
                'prob': prob,
                'odds': odd,
                'edge': edge,
                'ev': ev
            })
    
    return value_bets


# ============================================================================
# PHASE 7: ELITE BACKTEST
# ============================================================================

def run_elite_backtest(df, feature_cols, split_date="2024-01-01"):
    """
    Elite production backtest with:
    - Time-aware calibration
    - Fixed ensemble
    - Multi-market betting
    - CLV tracking
    - Portfolio approach
    """
    print("\n🏆 ELITE BACKTEST - ALL FIXES APPLIED")
    print("="*80)
    
    # Time split
    train_df = df[df['Date'] < split_date].copy()
    test_df = df[df['Date'] >= split_date].copy()
    
    print(f"Training: {len(train_df):,} matches")
    print(f"Testing:  {len(test_df):,} matches")
    
    # Train elite ML
    ml_model = train_elite_ml(train_df, feature_cols)
    
    # Train Dixon-Coles
    print("\n🎯 Training Elite Dixon-Coles...")
    dc_models = {}
    for league in train_df['League'].unique():
        dc_models[league] = EliteDixonColes(xi=0.002).fit(train_df, league)
    
    # Backtest
    print("\n📊 Running elite backtest...")
    
    bankroll = 10000
    peak = 10000
    trades = []
    clv_tracker = CLVTracker()
    
    for idx in tqdm(range(len(test_df)), desc="Testing"):
        row = test_df.iloc[idx]
        
        odds = np.array([row.get('AvgH'), row.get('AvgD'), row.get('AvgA')])
        
        if np.any(pd.isna(odds)) or np.any(odds <= 1.0):
            continue
        
        # Get features
        features = row[feature_cols].fillna(0).values.reshape(1, -1)
        
        # ML prediction
        ml_probs = ml_model.predict_proba(features)[0]
        
        # DC prediction
        try:
            dc_pred = dc_models[row['League']].predict_exact(row['HomeTeam'], row['AwayTeam'])
            dc_probs = np.array([dc_pred['prob_home'], dc_pred['prob_draw'], dc_pred['prob_away']])
        except:
            continue
        
        # Fixed ensemble
        ensemble_probs = ensemble_fixed(dc_probs, ml_probs, dc_weight=0.6)
        
        # Find ALL value bets
        value_bets = find_all_value_bets(ensemble_probs, odds, min_ev=0.03, min_prob=0.55)
        
        if not value_bets:
            continue
        
        # Place bets
        for bet in value_bets:
            prob = bet['prob']
            odd = bet['odds']
            
            # Kelly
            kelly = (prob * (odd - 1) - (1 - prob)) / (odd - 1)
            kelly = max(0, kelly) * 0.20
            kelly = min(kelly, 0.02)
            
            stake = bankroll * kelly
            
            # Outcome
            outcomes = ['H', 'D', 'A']
            won = row['FTR'] == outcomes[bet['market_idx']]
            
            profit = stake * (odd - 1) if won else -stake
            bankroll += profit
            peak = max(peak, bankroll)
            
            # CLV (if closing odds available)
            closing_col = {'H': 'AvgCH', 'D': 'AvgCD', 'A': 'AvgCA'}.get(outcomes[bet['market_idx']])
            if closing_col and closing_col in row:
                closing_odd = row.get(closing_col)
                if not pd.isna(closing_odd):
                    clv = clv_tracker.calculate_clv(odd, closing_odd)
                    clv_tracker.record(clv)
            
            trades.append({
                'date': row['Date'],
                'match': f"{row['HomeTeam']} vs {row['AwayTeam']}",
                'market': outcomes[bet['market_idx']],
                'prob': prob,
                'odds': odd,
                'ev': bet['ev'],
                'stake': stake,
                'won': won,
                'profit': profit,
                'bankroll': bankroll
            })
    
    # Results
    trades_df = pd.DataFrame(trades)
    
    if len(trades_df) == 0:
        print("❌ No trades")
        return None, None
    
    clv_stats = clv_tracker.get_stats()
    
    stats = {
        'initial': 10000,
        'final': bankroll,
        'profit': bankroll - 10000,
        'roi_pct': ((bankroll - 10000) / 10000) * 100,
        'n_trades': len(trades_df),
        'win_rate': trades_df['won'].mean() * 100,
        'avg_prob': trades_df['prob'].mean() * 100,
        'avg_ev': trades_df['ev'].mean() * 100,
        'max_dd': ((peak - trades_df['bankroll'].min()) / peak) * 100,
        'mean_clv': clv_stats['mean_clv'],
        'positive_clv_pct': clv_stats['positive_pct']
    }
    
    print("\n" + "="*80)
    print("🏆 ELITE BACKTEST RESULTS")
    print("="*80)
    print(f"Initial:         ${stats['initial']:,.2f}")
    print(f"Final:           ${stats['final']:,.2f}")
    print(f"Profit:          ${stats['profit']:+,.2f}")
    print(f"ROI:             {stats['roi_pct']:.2f}%")
    print(f"Trades:          {stats['n_trades']}")
    print(f"Win Rate:        {stats['win_rate']:.1f}%")
    print(f"Avg Prob:        {stats['avg_prob']:.1f}%")
    print(f"Avg EV:          {stats['avg_ev']:.2f}%")
    print(f"Max Drawdown:    {stats['max_dd']:.2f}%")
    print(f"Mean CLV:        {stats['mean_clv']:.2f}%")
    print(f"Positive CLV:    {stats['positive_clv_pct']:.1f}%")
    print("="*80)
    
    return trades_df, stats


# ============================================================================
# EXECUTE ELITE SYSTEM
# ============================================================================

print("\n🎯 Loading data...")
df_raw = pd.read_csv("https://www.football-data.co.uk/mmz4281/2324/E0.csv", encoding='latin1')
df_raw['Date'] = pd.to_datetime(df_raw['Date'], dayfirst=True)
df_raw = df_raw.sort_values('Date')
df_raw['League'] = 'Premier League'
df_raw['FTHG'] = df_raw['FTHG'].astype(int)
df_raw['FTAG'] = df_raw['FTAG'].astype(int)
df_raw['Outcome'] = df_raw['FTR'].map({'H': 0, 'D': 1, 'A': 2})
df_raw['DaysSinceMatch'] = (df_raw['Date'].max() - df_raw['Date']).dt.days

print(f"✅ Loaded {len(df_raw):,} matches")

print("\n🔧 Engineering elite features...")
df, feature_cols = create_elite_features(df_raw)
print(f"✅ Ready: {len(df):,} matches, {len(feature_cols)} features")

print("\n🏆 Running elite backtest...")
trades_df, stats = run_elite_backtest(df, feature_cols, split_date="2024-01-01")

if trades_df is not None:
    plt.figure(figsize=(14, 6))
    plt.plot(trades_df['bankroll'].values, linewidth=2, color='green' if stats['roi_pct'] > 0 else 'red')
    plt.axhline(10000, color='blue', linestyle='--', alpha=0.5)
    plt.title(f"Elite System - ROI: {stats['roi_pct']:.2f}% | CLV: {stats['mean_clv']:.2f}%", 
              fontsize=16, fontweight='bold')
    plt.xlabel('Trade Number')
    plt.ylabel('Bankroll ($)')
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()
    
    print("\n✅ ELITE SYSTEM READY FOR PRODUCTION!")
```
