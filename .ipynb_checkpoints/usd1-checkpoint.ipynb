```python
stocks = pd.read_csv("stocksticker" , header = [0,1], index_col = [0], parse_dates = [0])
stocks
```

```python
stocks.columns
```

```python
#convert multiindex to one tuple
stocks.columns = stocks.columns.to_flat_index()
stocks.columns
```

```python
stocks
```

```python
stocks.columns = pd.MultiIndex.from_tuples(stocks.columns)# convert tuples to multi index
stocks
```

```python
stocks.describe()
```

```python
close = stocks.loc[:,'Close'].copy() # copy the details to close
```

```python
close
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
```

```python
ax = close.plot(figsize=(12,6), fontsize = 12)
ax.set_title("Closing Prices for AAPL KO SPY")
ax.set_xlabel("Date")
ax.set_ylabel("price")
plt.tight_layout()
plt.legend(fontsize = 12)
plt.show() # forces render to show the grid to show
```
