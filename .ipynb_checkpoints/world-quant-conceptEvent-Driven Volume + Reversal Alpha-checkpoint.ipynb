## WORLDQUANT → PYTHON RESEARCH PIPELINE (Event-Driven Volume + Reversal Alpha)

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("seaborn-v0_8")
```

## Download Data (HANDLE MULTIINDEX PROPERLY)

```python
def download_data(tickers, start, end):
    raw = yf.download(tickers, start=start, end=end, auto_adjust=True)
    
    df = (
        raw
        .stack(level=1, future_stack=True)
        .reset_index()
        .rename(columns={"level_1": "Ticker"})
        .sort_values(["Ticker", "Date"])
        .reset_index(drop=True)
    )
    
    print(f"Shape: {df.shape}")
    
    # Try display (Jupyter), fall back to print
    try:
        display(df.head(15))
    except NameError:
        print(df.head(15).to_string())
    
    return df
```

```python
tickers = ["AAPL", "MSFT", "KO", "PEP", "XOM"]
df = download_data(tickers, "2020-01-01", "2024-01-01")
```

## Feature Engineering (WorldQuant Operators → Python)
## | WorldQuant        | Python                         |
## | ----------------- | ------------------------------ |
## | `ts_delta(x,n)`   | `x - x.shift(n)`               |
## | `ts_sum(x,n)`     | `x.rolling(n).sum()`           |
## | `ts_arg_max(x,n)` | `rolling().apply(np.argmax)`   |
## | `rank(x)`         | `groupby(Date).rank(pct=True)` |


```python
def add_features(df):
    df = df.copy()

    # Returns
    df["returns"] = df.groupby("Ticker")["Close"].pct_change()

    # 2-day price delta (ts_delta)
    df["delta_2"] = df.groupby("Ticker")["Close"].diff(2)

    # 5-day volume average
    df["vol_avg_5"] = df.groupby("Ticker")["Volume"].rolling(5).mean().reset_index(0, drop=True)

    # ts_arg_max(volume, 5)
    df["vol_argmax_5"] = (
        df.groupby("Ticker")["Volume"]
        .rolling(5)
        .apply(lambda x: np.argmax(x.values), raw=False)
        .reset_index(0, drop=True)
    )

    return df
```

```python
df_feat = add_features(df)
```

## Trigger Logic (trade_when equivalent)
## trigger = ts_arg_max(volume,5) < 1 && volume >= avg(volume,5)
## alpha   = -rank(ts_delta(close,2))


```python
def build_alpha(df):
    df = df.copy()

    # Trigger condition
    df["trigger"] = (
        (df["vol_argmax_5"] < 1) &
        (df["Volume"] >= df["vol_avg_5"])
    ).astype(int)

    # Alpha expression
    df["alpha_raw"] = -df["delta_2"]

    # Cross-sectional rank by date
    df["alpha_rank"] = (
        df.groupby("Date")["alpha_raw"]
        .rank(pct=True)
    )

    # trade_when gate
    df["alpha_final"] = np.where(df["trigger"] == 1, df["alpha_rank"], 0)

    return df
```

```python
df_alpha = build_alpha(df_feat)
```

## Clean NaNs & Sanity Tables

```python
df_clean = (
    df_alpha
    .dropna(subset=["alpha_final", "Close", "Volume","returns","delta_2","vol_avg_5","vol_argmax_5","alpha_raw","alpha_rank"])
    .reset_index(drop=True)
)

print("ALPHA SNAPSHOT:")
display(df_clean.head(10))
```

```python
print("Trigger distribution:")
print(df_clean["trigger"].value_counts())
```

## VISUALIZATIONS

## A. Price + Volume Sanity Check (Single Ticker)


```python
def visualize_price_volume(df, ticker, start, end):
    sub = df[
        (df["Ticker"] == ticker) &
        (df["Date"] >= start) &
        (df["Date"] <= end)
    ]

    fig, ax1 = plt.subplots(figsize=(14,6))

    ax1.plot(sub["Date"], sub["Close"], color="black", label="Close Price")
    ax1.set_ylabel("Price")
    ax1.legend(loc="upper left")

    ax2 = ax1.twinx()
    ax2.bar(sub["Date"], sub["Volume"], alpha=0.3, label="Volume")
    ax2.set_ylabel("Volume")
    ax2.legend(loc="upper right")

    plt.title(f"{ticker} — Price & Volume (Sanity Check)")
    plt.show()
```

```python
visualize_price_volume(df_clean, "AAPL", "2021-01-01", "2025-12-31")
```

## B. Trigger Overlay (When Alpha Is Active)

```python
def visualize_triggers(df, ticker):
    sub = df[df["Ticker"] == ticker]

    plt.figure(figsize=(14,6))
    plt.plot(sub["Date"], sub["Close"], label="Price", color="black")

    plt.scatter(
        sub[sub["trigger"] == 1]["Date"],
        sub[sub["trigger"] == 1]["Close"],
        color="red",
        label="Trigger ON",
        s=20
    )

    plt.title(f"{ticker} — Trigger Days Highlighted")
    plt.legend()
    plt.show()
```

```python
visualize_triggers(df_clean, "AAPL")
```

## C. Portfolio-Level Alpha Signal (Cross-Sectional Mean)

```python
portfolio_alpha = (
    df_clean
    .groupby("Date")["alpha_final"]
    .mean()
    .cumsum()
)

plt.figure(figsize=(14,6))
plt.plot(portfolio_alpha, label="Alpha Signal (Cumulative)")
plt.title("Alpha Signal — Portfolio Level")
plt.legend()
plt.show()
```
