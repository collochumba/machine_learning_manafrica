```python
# ============================================
# WorldQuant Alpha Scanner (Fixed)
# ============================================

import yfinance as yf
import pandas as pd
import numpy as np
from tqdm import tqdm

# --------------------------------------------
# Top 100 stocks
# --------------------------------------------

tickers = [
"AAPL","MSFT","AMZN","NVDA","GOOGL","META","TSLA","BRK-B","UNH","JNJ",
"XOM","JPM","V","PG","MA","HD","CVX","LLY","ABBV","MRK",
"PEP","KO","COST","AVGO","ADBE","WMT","BAC","MCD","TMO","ACN",
"ABT","LIN","NFLX","DHR","AMD","CMCSA","ORCL","NKE","TXN","INTC",
"PM","UNP","LOW","HON","UPS","RTX","QCOM","IBM","SBUX","MS",
"CAT","INTU","GS","SPGI","BLK","AMGN","PLD","DE","MDT","ISRG",
"NOW","ADP","SYK","LMT","AXP","GE","TJX","CB","GILD","C",
"AMT","BKNG","T","ZTS","MMC","MO","USB","PGR","SCHW","DUK",
"SO","BDX","EL","CL","CSX","SLB","REGN","FIS","ITW","PNC",
"VRTX","ADI","ETN","ICE","EOG","KLAC","SHW","PSA","AON","APD"
]

start = "2018-01-01"

results = []

# --------------------------------------------
# Alpha Calculation
# --------------------------------------------

def compute_alpha(df):

    typical = (df["High"] + df["Low"] + df["Close"]) / 3
    df["vwap"] = (typical * df["Volume"]).cumsum() / df["Volume"].cumsum()

    signal = -((df["vwap"]/df["Open"]) * ((df["High"]/df["Low"]) + 100))

    df["alpha"] = signal.rolling(2).mean()

    return df


# --------------------------------------------
# Strategy Evaluation
# --------------------------------------------

def evaluate(df):

    df["future_return"] = df["Close"].shift(-1) / df["Close"] - 1

    threshold = df["alpha"].quantile(0.2)

    df["signal"] = (df["alpha"] < threshold).astype(int)

    strat = df["signal"] * df["future_return"]

    strat = strat.dropna()

    if strat.std() == 0 or len(strat) < 50:
        return None

    sharpe = np.sqrt(252) * strat.mean() / strat.std()

    annual_return = strat.mean() * 252

    return sharpe, annual_return


# --------------------------------------------
# Run Scanner
# --------------------------------------------

for ticker in tqdm(tickers):

    try:

        df = yf.download(ticker, start=start, progress=False)

        if df.empty or len(df) < 200:
            continue

        df = compute_alpha(df)

        result = evaluate(df)

        if result is None:
            continue

        sharpe, returns = result

        results.append({
            "Ticker": ticker,
            "Sharpe": sharpe,
            "AnnualReturn": returns
        })

    except:
        continue


# --------------------------------------------
# Results
# --------------------------------------------

results_df = pd.DataFrame(results)

if len(results_df) == 0:
    print("No valid results found")
else:

    results_df = results_df.sort_values("Sharpe", ascending=False)

    print("\nTop Stocks For This Alpha\n")
    print(results_df.head(20))
```
